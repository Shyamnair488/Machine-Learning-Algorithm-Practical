Import Librraries :

```python
import numpy as np
import matplotlib.pyplot as mtp
import pandas as pd
```

```python
dataset1 = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/mla/suv_data - suv_data - suv_data - suv_data.csv")
dataset1
```

```python
dataset=dataset1.dropna()
```

```python
x = dataset.iloc[:,[2,3]].values
print(x)
y = dataset.iloc[:,4].values
print(y)
```

Splitting Dataset into Training & Testing set :

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.25,random_state=3)
```

Feature Scaling :

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
x_train = sc.fit_transform(x_train)
x_test = sc.transform(x_test)
```

Fitting Naive Bayes to the Training Set :

```python
from sklearn.naive_bayes import GaussianNB
classifier = GaussianNB()
```

```python
classifier.fit(x_train,y_train)
```

Predicting the Test set results :

```python
y_pred = classifier.predict(x_test)
```

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test,y_pred)
```

Visualization :

```python
from matplotlib.colors import ListedColormap
x_set,y_set = x_train,y_train
X1,X2 = np.meshgrid(np.arange(start = x_set[:,0].min()-1,stop = x_set[:,0].max()+1,step=0.01),
np.arange(start=x_set[:,1].min()-1,stop=x_set[:,0].max()+1,step=0.01))
mtp.contourf(X1,X2,classifier.predict(np.array([X1.ravel(),X2.ravel()]).T).reshape(X1.shape),
             alpha=0.75,cmap=ListedColormap(('white','green')))
mtp.xlim(X1.min(),X1.max())
mtp.ylim(X2.min(),X2.max())
for i, j in enumerate(np.unique(y_set)):
  mtp.scatter(x_set[y_set==j,0],x_set[y_set==j,1],
              c=ListedColormap(('red','cyan'))(i),label =j)
  mtp.title("Naive Bayes ( Training Set )")
  mtp.xlabel('Age')
  mtp.ylabel('Estimated Salary')
  mtp.legend()
```
