# **SVM**

```python
import numpy as np
import pandas as pd
```

```python
from sklearn import svm
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
```

```python
# Loading the iris dataset
iris = load_iris()
```

```python
# Splitting the dataset into train and test sets
X_train, X_test, y_train, y_test = train_test_split(iris.data, iris.target, test_size=0.5, random_state=42)
```

```python
from sklearn.svm import SVC # "Support vector classifier"
cfl = SVC(kernel='linear')
cfl.fit(X_train, y_train)
```

```python
y_pred = cfl.predict(X_test)
```

```python
# Calculating the accuracy of the model
accuracy = accuracy_score(y_test, y_pred)

print("Accuracy of the SVC model:", accuracy)
```
